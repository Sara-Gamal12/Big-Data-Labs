```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark.sql.types import StructType, StructField, IntegerType

```

```python
spark=SparkSession.builder\
    .master("local[*]")\
    .appName("MapReduce")\
    .getOrCreate()
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("card", IntegerType(), True)
])
df = spark.read.csv("/in.csv", header=False, schema=schema)
```

```python
result = df.groupBy("id").sum("card")
```

```python
result=result.orderBy("id", ascending=True)
```

```python
#show first 100
result.show(100)
```

```python
#show last 100
last_100=result.orderBy("id", ascending=False).limit(100)
last_100.orderBy("id", ascending=True).show(100)
```
